### How to use two(2) axes in one canva or figure

```python
# Create blank Canvas or figure
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,2,2])#main axes
axes2=fig.add_axes([0.2,0.5,0.4,0.3])#insert axes
axes3=fig.add_axes([1.2,1.5,0.5,0.5])#insert axes
axes1.plot(x,y,'g')#ploting on axes1
axes2.plot(x,y,'r')#ploting on axes2
axes3.plot(x,y,'b')#ploting on axes2
# setting x and ylabels for the two axes or graphs
axes1.set_xlabel('set xlabel1')
axes1.set_ylabel('set ylabel1')
axes1.set_title("set title1")
axes2.set_xlabel('set xlabel2')
axes2.set_ylabel('set ylabel2')
axes2.set_title("set title2")
fig.suptitle('Vertically stacked subplots')
```

```python
#creating a subplot
fig,axes=plt.subplots()
axes.plot(x,y,'r')
axes.set_xlabel('set xlabel')
axes.set_ylabel('set ylabel')
axes.set_title("set title")
```

#### Using label to specify the plane on the graph

```python
#creating a subplot
fig,ax=plt.subplots(figsize=(5, 2.7),)
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend(loc=4);
```

```python
fig, (ax1, ax2,ax3,ax4) = plt.subplots(1,4)
fig.suptitle('Horizontally stacked subplots')
ax1.plot(x, y)
ax2.plot(x, -y)
ax3.plot(-x, -y)
ax4.plot(-x, y)
```

```python
fig, axs = plt.subplots(2,2)
fig.suptitle('Horizontally stacked subplots')
axs[0,0].plot(x, y)
axs[0,0].set_title('Axis [0, 0]')
axs[0,1].plot(x, -y)
axs[0,1].set_title('Axis [0, 1]')
axs[1,0].plot(-x, -y)
axs[1,0].set_title('Axis [1, 0]')
axs[1,1].plot(-x, y)
axs[1,1].set_title('Axis [1,1]')

for ax in axs.flat:
    ax.set(xlabel='x-label',ylabel='y-lable')
for ax in axs.flat:
    ax.label_outer()
```

```python
x = np.linspace(0, 2, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(5, 2.7))
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend(loc=1);  # Add a legend.
# fig.savefig('figure_lengend.png',dpi=200)
```
